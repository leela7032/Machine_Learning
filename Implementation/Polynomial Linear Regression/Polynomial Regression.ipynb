# Polynomial Regression

## Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the Dataset

```python
Data=pd.read_csv("RBC Polynomial.csv")
X=Data.iloc[:,0:-1].values
Y=Data.iloc[:,-1].values
```

```python
plt.scatter(X,Y)
```

Here we observe the data points behaviour so that we can analyze which algorithm can be used ,so here the points are in non-Linear fashion.

## Training with Linear Regression for whole dataset.

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
X1=reg.fit(X,Y)
print(X1)
```

## Training with Polynomial Regression for the given dataset with degree 6.

```python
from sklearn.preprocessing import PolynomialFeatures
Poly_reg=PolynomialFeatures(degree =6)
X_poly=Poly_reg.fit_transform(X)

X2=LinearRegression()
X_Poly=X2.fit(X_poly,Y)
print(X_Poly)

```

## Visualizing the Linear Regression result

```python
plt.scatter(X,Y)
plt.plot(X,reg.predict(X),color="Red")
plt.title("RBC-Polynomial Regression",fontsize=20)
plt.xlabel("Age", fontsize=20)
plt.ylabel("RBC's",fontsize=20)
plt.show()
```

## Visualizing the Polynomial Regression with degree 6.

```python
plt.scatter(X,Y)
plt.plot(X,X2.predict(X_poly),color="Red")
plt.title("RBC - Linear Regression")
plt.xlabel('Age')
plt.ylabel('RBC')
plt.show()
```

## Predicting a new result with Linear Regression

```python
reg.predict([[6.5]])
```

## Predicting a new result with Polynomial Linear Regression

```python
X2.predict(Poly_reg.fit_transform([[6.5]]))
```
