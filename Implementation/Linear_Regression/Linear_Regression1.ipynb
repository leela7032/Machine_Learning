# Linear Regression

##         Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Here we use these Python Libraries for our algorithm implementation

# Importing DataSet

```python
Lime=pd.read_csv("SalaryPrediction.csv")
X=Lime.iloc[:,:-1].values
Y=Lime.iloc[:,-1].values
```

```python
print(X)
```

```python
print(Y)
```

# Splitting the Data Set into Train and Test data sets

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.8,random_state=1)
```

Here random state means that the splitting happens for the dataset everytime so if we keep randomstate as any number then even if we type this number we get exact splitting as before. 

If we don't mention then everytime we run it will take random splitting takes place.

# Training the DataSet with Linear Regression

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,Y_train)
```

# Predicting the output for X_test

```python
Y_pred=reg.predict(X_test)
print(Y_pred)
```

# Visualize the Training Set

```python
plt.scatter(X_train,Y_train,color="red")
plt.plot(X_train,reg.predict(X_train),color="blue")
plt.title("Prediction of CTC Of an employee")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

# Visualizing the Testing Set

```python
plt.scatter(X_test,Y_test,color="red")
plt.plot(X_train,reg.predict(X_train),color="blue")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Prediction of salary based on experience")
plt.show()
```

# Accuracy Prediction (or) Performance

```python
from sklearn.metrics import r2_score
r2_score(Y_pred,Y_test)
```
